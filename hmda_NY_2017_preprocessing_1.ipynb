# 1) Iteration by chunks: Null values
    The dataset is massive so we will deal with it by chunks.
    We split the dataset in chunks of x size, then iterate by each chunk and finally we summarize the total column's null values.

```python
import pandas as pd
```

```python
# We force to show all the records
pd.set_option('display.max_rows', None) 

# We create a dictionary to keep all null values from the different chunks
total_null_values = None

# Iterate across the different chunks directly from the dataset
for chunk in pd.read_csv('C:/Users/nonox/hmda_2017_ny.csv', chunksize=100000):
    chunk_null_values = chunk.isnull().sum()
    
    if total_null_values is None:
        total_null_values = chunk_null_values
    else:
        total_null_values += chunk_null_values
    
    # Total null values per chunk
    print(f'Null values per column in this chunk:\n{chunk_null_values}\n')

# Total null values for the whole dataset
print(f'Total null values per column in the entire dataset:\n{total_null_values}')
```

# 3) Iteration by chunks: "No co-applicants"
    a) First step in our data selection is to try to reduce the number of record as soon as possible.
    b) By focusing on "No co-aplicants" we significantly reduce the number of records.
    c) Our focus is on single individuals, we only take "No co-applicat" records.

```python
# Initialize an empty DataFrame to store the filtered data
filtered_df = pd.DataFrame()

# Processing the CSV file in chunks
for chunk in pd.read_csv('C:/Users/nonox/hmda_2017_ny.csv', chunksize=100000):
    # Filter the chunk where "co_applicant_ethnicity_name" is "No co-applicant"
    filtered_chunk = chunk[chunk['co_applicant_ethnicity_name'] == 'No co-applicant']
    
    # Append the filtered chunk to the filtered_df DataFrame
    filtered_df = pd.concat([filtered_df, filtered_chunk])

# New CSV file
filtered_df.to_csv('C:/Users/nonox/hmda_2017_ny_no_co_applicant.csv', index=False)

print(f"Filtered dataset saved to '{'C:/Users/nonox/hmda_2017_ny_no_co_applicant.csv'}'")


#https://www.geeksforgeeks.org/how-to-create-an-empty-dataframe-and-append-rows-columns-to-it-in-pandas/
```

# 6) Filtering columns:

#### Looking at "action_taken" column:
###### Our target is approved/denied. Therefore we will keep only:

1) Loan originated                                        111541
2) Application denied by financial institution             32229
3) Application approved but not accepted                    9555

###### Hence, from "action_taken" column we will get rid off:
1) Application withdrawn by applicant                      28299
2) Loan purchased by the institution                       14734
3) File closed for incompleteness                          11915
4) Preapproval request denied by financial institution         5
5) Preapproval request approved but not accepted               1

#### From the column "denial_reason_name_1":
1) We will get rid of "Credit application incomplete"

```python
df = df[(df['denial_reason_name_1'] != "Credit application incomplete") &

        (df['action_taken_name'] != "Application withdrawn by applicant") &
        (df['action_taken_name'] != "Loan purchased by the institution") & 
        (df['action_taken_name'] != "File closed for incompleteness") &
        (df['action_taken_name'] != "Preapproval request denied by financial institution") &
        (df['action_taken_name'] != 'Preapproval request approved but not accepted')]
```
